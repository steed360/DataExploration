# Create Transactions

```python
dfTransactions = dfClean[['Quantity', 'InvoiceDate', 'SalesAmount', 'StockCode', 'Customer ID']]
dfTransactions.head()
```

# Create Customer Dimension

```python
dfCustomers = dfClean.groupby('Customer ID').agg({'Quantity': 'sum', 'SalesAmount': 'sum'}).reset_index()

#dfCustomers.head()
```

```python
dfCustomers['SalesAmountDecile'] = pd.qcut(dfCustomers['SalesAmount'], 10,  labels = False)
```

```python
# Bin the SalesAmount column
bins = [600000,200000,100000,50000,20000,10000,5000,1000,500,0,-100000000]
bins.reverse()
labels = ['200,001-600,000','100,001-200,001','50,001-100,000','20,000-100,000','10,000-20,000','5,000-10,000','1,001-5,000','501-1000','0-500','Under Zero'   ]
labels.reverse()
dfCustomers['SalesAmountBanded'] =pd.cut( dfCustomers['SalesAmount'] , bins,  labels=labels)

# Check the bandings
dfCustomers.groupby('SalesAmountBanded').agg({'SalesAmount': ['min','max']}).reset_index()
```

# DIM Countries

```python
dfCountryBandings =  dfClean.groupby('Country').agg({'Quantity': 'sum', 'SalesAmount': 'sum'}).reset_index()
dfCountryBandings.sort_values ('SalesAmount', ascending=False)

```
